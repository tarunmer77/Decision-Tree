```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
data = pd.read_csv("Company_Data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# visualizing Sales
```

```python
import seaborn as sns
```

```python
sns.boxplot(data['Sales'])
```

```python
# as we can see that maximum sale is in between 5-9 
# let's categorize sales as High if it is  >=7.5 and low if <7.5
```

```python
sales_val = []
for value in data['Sales']:
    if value >= 7.5:
        sales_val.append('High')
    else:
        sales_val.append('Low')
data['Sales_category'] = sales_val
```

```python
data.head()
```

```python
data.columns
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
data['US'] = label_encoder.fit_transform(data['US'])
data['Urban'] = label_encoder.fit_transform(data['Urban'])
data['ShelveLoc'] = label_encoder.fit_transform(data['ShelveLoc'])
```

```python
data.head(10)
```

```python
data.shape
```

```python
data.info()
```

```python
colnames = list(data.columns)
```

```python
colnames
```

```python
# splitting data into test and train
```

```python
x = data.iloc[:,1:11]
```

```python
x.head()
```

```python
y = data['Sales_category']
```

```python
y.head()
```

```python
y.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
# model building
```

```python
# Model 1
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python

#PLot the decision tree
tree.plot_tree(model);
```

```python
x.columns
```

```python

fn= ['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']
cn=['High','Low']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

```python
# prediction
```

```python
pred = model.predict(x_test)
pd.Series(pred).value_counts()
```

```python
pred
```

```python
pd.crosstab(y_test,pred)
```

```python
np.mean(pred==y_test)
```

```python
# accuracy = 65%
```

```python
# Model 2
```

```python
model2 = DecisionTreeClassifier(criterion = 'entropy',max_depth=5)
model2.fit(x_train,y_train)
fn= ['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']
cn=['High','Low']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(model2,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

```python
pred2 = model2.predict(x_test)
pd.Series(pred2).value_counts()
```

```python
np.mean(pred2==y_test)
```

```python
# accuracy increased to 67%
```

```python
# Model 3
```

```python
# categorizing Sales value as High , Mid and Low
```

```python
sales_val2 = []
for value in data['Sales']:
    if value >= 9:
        sales_val2.append('High')
    elif value >= 6 and value < 9:
        sales_val2.append('Mid')
    else:
        sales_val2.append('Low')
data['Sales_category2'] = sales_val2
```

```python
data.head()
```

```python
x2 = data.iloc[:,1:11]
y2 = data['Sales_category2']
```

```python
x2.head()
```

```python
y2.head()
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(x2,y2,test_size=0.3,random_state=40)
```

```python
model3 = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model3.fit(x2_train,y2_train)
```

```python

fn2= ['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']
cn2=['High','Mid','Low']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(model3,
               feature_names = fn2, 
               class_names=cn2,
               filled = True);
```

```python
pred3 = model3.predict(x_test)
pd.Series(pred3).value_counts()
```

```python
np.mean(pred3==y2_test)
```

```python
# here accuracy is 53%
```
